```python
import random
```

```python
def fillSetX():
    p = random.randint(1000,3000)
    #print(p)

    X = set()
    #print(X)

    for i in range(p):
        num = random.randint(-3000,3000)

        while num in X:
            num = random.randint(-3000,3000)

        if num not in X:
            X.add(num)

    #print(X)
    print("Set X contains",p,"integers.")
    return X
```

```python
def fillSetY():
    q = random.randint(500,1000)
    #print(q)

    Y = set()
    #print(Y)

    for i in range(q):
        num = random.randint(-3000,3000)

        while num in Y:
            num = random.randint(-3000,3000)

        if num not in Y:
            Y.add(num)

    #print(Y)
    print("Set Y contains",q,"integers.")
    return Y
```

```python
def fillSetZ():
    r = random.randint(500,1000)
    #print(r)

    Z = set()
    #print(Z)

    for i in range(r):
        num = random.randint(-3000,3000)

        while num in Z:
            num = random.randint(-3000,3000)

        if num not in Z:
            Z.add(num)

    #print(Z)
    print("Set Z contains",r,"integers.")
    return Z
```

```python
class node:
    def __init__(self, value = None):
        self.value = value
        self.left = None
        self.right = None
        self.parent = None
```

```python
class binary_search_tree:
    def __init__(self):
        self.root = None
        self.comparisons = 0
    
    def insert(self, value):
        #Check if root node is filled yet
        if self.root is None:
            self.root = node(value)
            #print(f"Inserted {value} as root")
        
        #If root already defined move onto next insert function
        else:
            self._insert(value, self.root)
            
    def _insert(self, value, temp_node):
        self.comparisons += 1
        #Check if past value is less than the value inside the current node
        if value < temp_node.value:
            #If current node doesn't have a left child, insert it
            if temp_node.left is None:
                temp_node.left = node(value)
                temp_node.left.parent = temp_node
                #print(f"Inserted {value} as left child of {temp_node.value}")
                
            #Otherwise pass the child into the function recursively
            else:
                self._insert(value, temp_node.left)
        
        #Some operation as above bit when the past value is more than the value inside the current node
        if value > temp_node.value:
            if temp_node.right is None:
                temp_node.right = node(value)
                temp_node.right.parent = temp_node
                #print(f"Inserted {value} as right child of {temp_node.value}")
                
            #Otherwise pass the child into the function recursively
            else:
                self._insert(value, temp_node.right)
       
    #Functions to get height           
    def height(self):
        return self._height(self.root)
        
    def _height(self, temp_node):
        if temp_node is None: 
            return -1
        else:
            left_height = self._height(temp_node.left)
            right_height = self._height(temp_node.right)
            return max(left_height,right_height) + 1
        
    #Search functions (for node)
    def find(self, value):
        self.comparisons += 1
        
        if self.root is None:
            return False
        else:
            return self._find(value, self.root)
        
    def _find(self, value, temp_node):
        if value == temp_node.value:
            return temp_node
        elif value < temp_node.value and temp_node.left is not None:
            return self._find(value, temp_node.left)
        elif value > temp_node.value and temp_node.right is not None:
            return self._find(value, temp_node.right)
    
    #Functions to delete
    def delete_value(self, value):
        return self.delete_node(self.find(value))
    
    def delete_node(self, temp_node):
        
        #gets node with lowest value using input node as root
        def min_node(n):
            temp = n
            while temp.left is not None:
                temp = temp.left
            return temp
        
        #gets number of children of the node
        def child_count(n):
            child_count = 0
            if n.left is not None: 
                child_count = child_count + 1
            if n.right is not None: 
                child_count = child_count + 1
            return child_count
        
        #parent of the node to be deleted
        parent_node = temp_node.parent
        
        #number of children of node to be deleted
        child_nodes = child_count(temp_node)
        
        #if node has no children
        if child_nodes == 0:
            if parent_node.left == temp_node:
                parent_node.left = None
            else:
                parent_node.right = None
        
        #if node has one child
        if child_nodes == 1:
            #get child node
            if temp_node.left is not None:
                child = temp_node.left
            else:
                child = temp_node.right
                
            #replace parent node with childe
            if parent_node.left == temp_node:
                parent_node.left = child
            else:
                parent_node.right = child
            
            #correct parent pointer
            child.parent = parent_node
            
        #if node has two children
        if child_nodes == 2:
            
            successor = min_node(temp_node.right)
            temp_node.value = successor.value
            
            self.delete_node(successor)
        
    #Search Functions (for value)
    def search(self, value):
        self.comparisons += 1
        
        if self.root is None:
            return False
        else:
            return self._search(value, self.root)
        
    def _search(self, value, temp_node):
        if temp_node is None:
            return False
        elif value == temp_node.value:
            return True
        elif value < temp_node.value and temp_node.left is not None:
            return self._search(value, temp_node.left)
        elif value > temp_node.value and temp_node.right is not None:
            return self._search(value, temp_node.right)
        
    #Functions to get the number of nodes:
    def node_count(self):
        return self._node_count(self.root)
    
    def _node_count(self, temp_node):
        if temp_node is None:
            return 0
        else:
            return self._node_count(temp_node.left) + self._node_count(temp_node.right) + 1
            
    #Print Functions  
    def print_tree(self):
        if self.root is not None:
            self._print_tree(self.root)
    
    def _print_tree(self, temp_node):
        if temp_node is not None:
            self._print_tree(temp_node.left)
            print(str(temp_node.value))
            self._print_tree(temp_node.right)
    
    #Function to count number of comparisons
    def count_comparisons(self):
        return self.comparisons
```

```python
def CalcIntersections(set1,set2):
    intersection= set1.intersection(set2) 
    return intersection
```

```python
#main cell

#function calls to fill the sets
X = fillSetX()
Y = fillSetY()
Z = fillSetZ()

#function calls to calculate intersections of sets
intersectionXY = CalcIntersections(X,Y)
intersectionXZ = CalcIntersections(X,Z)

print("Sets X and Y have",(len(intersectionXY)),"values in common.")
print("Sets X and Z have",(len(intersectionXZ)),"values in common.")

#Tree functions are called from seperate cells so the sets are only changed when user reruns the main cell
#This was done for easier testing and debugging
```

```python
#Functions to fill, delete and search on the BST
BST = binary_search_tree()

#Insert Set X into BST Tree
for i in X:
    BST.insert(i)

#Get the number of Comparisons to build the BST
BST_ComparisonsCount1 = BST.count_comparisons()
print("#of comparrisions ", BST_ComparisonsCount1)
    
#BST.print_tree()

#Get height of BST Tree
BST_Height1 = BST.height()
print("Tree height ",BST_Height1)

#Get the number of Nodes of BST Tree
BST_NodeCount1 = BST.node_count()
print("Node count",BST_NodeCount1)

#Delete element of Y from the BST Tree
for i in Y:
    check = BST.search(i)
    if check == True:
        BST.delete_value(i) 
    
BST_Height2 = BST.height()
print("Tree height ",BST_Height2)


BST_NodeCount2 = BST.node_count()
print("Node count",BST_NodeCount2)

    
BST_ComparisonsCount2 = BST.count_comparisons() 
print("#of comparrisions ", BST_ComparisonsCount2)

#Search for every element of Z in the tree
NumbersFound_BST = 0
NumbersNotFound_BST = 0

for i in Z:
    check = BST.search(i)
    if check == True:
        NumbersFound_BST = NumbersFound_BST + 1
    else:
        NumbersNotFound_BST = NumbersNotFound_BST + 1

#print("# found, ",NumbersFound_BST)
#print("# not found, ",NumbersNotFound_BST)

#Get the number of Comparisons to search the BST for elements of Z
BST_ComparisonsCount3 = BST.count_comparisons()
print("#of comparrisions ", BST_ComparisonsCount3)


```
